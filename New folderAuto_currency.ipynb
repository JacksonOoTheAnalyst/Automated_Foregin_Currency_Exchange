```python
#Gather the parametersof the user

date = input("Please Enter the Today Date as 'YYYY-MM-DD' 'latest' : ")
base = print("Euro Exchange rate base if you want more \nPlese subricipte 'Basic-Plan' from 'Free' Mode")
curr = input("Base Currency want to Exchange to (currency) : ")
amou = float(input("Exchange Currency amount : " ))

#Construct the URL and sned a GET request to it

base_url = "http://api.exchangeratesapi.io/v1"
url = base_url + "/"+ date + "?access_key=e3368f60e8f06eefd41860837f0150a5" + "&symbols=" + curr
print (url)
response = requests.get(url)

# Displaying the error message, if something went wrong
if(response.ok is False):
    print("\nError {}:".format(response.status_code))
    print(response.json()['error'])

else:
    data = response.json()
    rate = data['rates'][curr]
        
    result = amou*rate
    
    print("\n{0} {1} is equal to {2} {3}, based upon exchange rates on {4}".format(amou,'Euro',result,curr,data['date']))
```
